## Linear Regression

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([15,20,25,37,40,45,48,50,55,61,64,67,70])
y = np.array([100,135,135,150,250,270,290,360,375,400,500,1000,1500])

print(f"x is independent variable : {x} \n y is dependent variable : {y}")
```

```python
xMean = np.mean(x)
yMean = np.mean(y)

print(f"x mean is : {xMean} \n y mean is : {yMean}")
```

```python
convXY = 0
varX = 0
n = len(x)
for i in range(n):
    convXY += (x[i] - xMean)* (y[i]-yMean)
    varX += (x[i]-xMean)**2

m = convXY/varX
print(convXY)
print(varX)
print(m)
```

```python
c = yMean - m*xMean
print(c)
```

```python
y1 = m*6 + c
print(f"Y1: {y1}")
```

```python
y2 = m*7 + c
print(f"Y1: {y2}")
```

```python
y3 = m*8 + c
print(f"Y1: {y3}")
```

```python
yPredict = m*x + c
print(yPredict)
```

```python
plt.scatter(x,y, color = "#e91e63", label = "Scatter")
plt.xlabel("Hours Studied")
plt.ylabel("Final Exam Score")
plt.scatter(x,yPredict, color = "#00b8d4", label = "line")
plt.legend()
plt.show()
plt.plot(x,y, label ="Scatter")
plt.plot(x,yPredict, label = "Line")
plt.legend()
plt.show()
```

```python
rNum = 0
rDen = 0

for i in range(n):
    rNum += (yPredict[i] - yMean)**2
    rDen += (y[i] - yMean)**2

r = rNum / rDen

print(f"R Square: {r}")
print(f"Accuracy is: {round(r*100)}%")
print(f"Error: {round(100-r*100)}%")
```
